```python
df.job_title_short
```

```python
job_counts = df.job_title_short.value_counts()
```

**BARCHART**

```python
job_counts = job_counts.sort_values(ascending=True)
plt.barh(job_counts.index,job_counts.values)
plt.show()
```

**CHART LABELLING**

```python
job_counts = job_counts.sort_values(ascending=False)
plt.title('postings by jobs title')
plt.ylabel('counts of job postings')
plt.xticks(rotation=45, ha = 'right')
plt.bar(job_counts.index,job_counts)
plt.show()
```

**plotting with pandas & matplotlib**

```python
job_counts.plot(kind='bar')
plt.title('postings by jobs title')
plt.ylabel('counts of job postings')
plt.xlabel(" ")
plt.xticks(rotation=45, ha = 'right')
plt.show()
```
